```python
import pandas as pd
```

```python
import nltk
```

```python
x=review1.review
y=review1.rating
```

### Sentiment Analysis with the NaiveBayesAnalyzer

```python
from textblob.sentiments import NaiveBayesAnalyzer
import nltk
for i in x:
    j=TextBlob(i, analyzer=NaiveBayesAnalyzer())          #Will take some time
    print(j.sentiment)
# blob = TextBlob(text, analyzer=NaiveBayesAnalyzer())
```

### Extracting the positive elements from the review to give a star based Naive Vayes Analyzer

```python
from textblob.sentiments import NaiveBayesAnalyzer
import nltk
list1=[]
for i in x:
    j=TextBlob(i, analyzer=NaiveBayesAnalyzer())
    a=j.sentiment.p_pos                                   #Will take some time
    list1.append(a)
```

#### Positivity rating based on the Naive Bayes Analyzer

```python
for i in range(1,16):
    print(f"Review {i} rating: ",list1[i-1]) #The list with the ratings given by the Naive Bayes Analyzer. 
```

### Converting ratings into 1-5 star to compare with the actual ratings.

```python
list2=[]
for i in list1:
    if (0.9<i<=1):
        j=5                        
        list2.append(j)
    if (0.7<i<=0.9):
        j=4
        list2.append(j)
    if (0.5<i<=0.7):
        j=3
        list2.append(j)
    if (0.3<i<=0.5):
        j=2
        list2.append(j)
    if (0.0<i<=0.3):
        j=1
        list2.append(j)
```

### Ratings in stars based on the analyzer

```python
for i in range(1,16):
    print(f"Review {i} rating: ",list2[i-1])
```

### Adding a new column of analyzer's rating in the dataframe as Bayes_rating

```python
review1['Bayes_Rating']=list2
z=review1.Bayes_Rating
```

```python
review1.head(15)
```

```python
import sklearn
accuracy = sklearn.metrics.accuracy_score(list(y), list(z))
print(accuracy)
```

#### The prediction score is 66.66 %.
We can see that the analyzer is not able to classify the reviews in moderate rating. It either takes the review as extreme positive or extreme negative. We can see that in the 4th review by Alex Abad-Santos and 12th review by David Edelstein.


### Confusion matrix

```python
from sklearn.metrics import confusion_matrix

confusion = confusion_matrix(y_true=y, y_pred=z)  #y is the actual value. z is the predicted value.

print('Confusion matrix:')
print(confusion)
```

### Consfusion matrix graph

```python
import pandas as pd
import seaborn as sns

confusion_df = pd.DataFrame(confusion, index=range(5), columns=range(5))  #5*5 is the shape of matrix for index and columns. 
axes = sns.heatmap(confusion_df, annot=True, cmap='nipy_spectral_r')
```

### Checking the similarity of each reviewer from movie 1 with every reviewer from movie 2.
Will give multiple warnings for each comparison. 

```python
list3=[]
dict1={}
for i in c:
    for j in d:
                                #Running this will give many warnings when performing similarity just like happened with professor.
        index_1=c.index(i)+1      
        index_2=d.index(j)+1           #Creates a new dictionary with similarity and reviewers index.
        k=i.similarity(j)
        list3.append(k)
        print(k)
        dict1.update({f"Review {index_1} & Review {index_2} ":f"{k:.2f}"})
```

```python
for i,j in dict1.items():
    if float(j)>0.90:                        #Review 4 and review 3
        print(i+': '+str(j))              #Review 9 and review 9 #Review 11 and review 1 #review 12 and review 14
                                             #Review 14 and review 8
```

### We have 5 same reviewers for both the movies. We can see that out of 5, it could identify 1 of the reviewers who commented on both.  
#### Considering the similarity>0.90 as the same reviewer, the analyzer has identified 4 pairs which could be same. Out of which we can see that only one identified is correct.

#### Printing the names who are common in giving reviews for both the movies. 

```python
name_list1=list(review1.Name)
name_list2=list(review2.Name)
for i in name_list1:
    for j in name_list2:
        if i==j:
            s=name_list1.index(i)+1
            t=name_list2.index(j)+1
            print(f"Reviewer {s} in movie 1 and reviewer {t} in movie 1: {i}")
```

```python
#Total Combinations:
total_comb=len(list3)
print(f'total combinations: {total_comb}')
```

### Predicted true which were same is for Jack Koyle and thus it couldn't predict the other 4.
#### Wrong Predictions = Not same but considered same (false positive) + Same but not considered same (false negative) =4 + 4 =8

#### True Predictions= 210-8=202

```python
wrong_predictions=8
true_predictions=202
total_predictions=210
percent_score=true_predictions/total_predictions
print(f'Score: {percent_score}')
```

### Thus the prediction score comes out to be 96.19%

### How reliable is this type of analysis? 
#### The predicition is more reliable when the text input given are long such as a book rather than small reviews. If considered the True Negative (Not same and not considered same), then only the prediction score comes out to be good.
